```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed set for reproducibility
np.random.seed(42)

# Number of samples
n = 100

# Generate X1 from N(3, 9)
X1 = np.random.normal(3, np.sqrt(9), n)

# Generate X2 from 0.5 * X1 + N(4, 4)
X2 = 0.5 * X1 + np.random.normal(4, np.sqrt(4), n)

# Combine X1 and X2 into a 2D array
samples = np.vstack((X1, X2)).T
```

```python
# Q1. Finding the mean in R^2 of the sample
mean_sample = np.mean(samples, axis=0)
print("Mean in R^2 of the sample :\n", mean_sample)
```

```python
# Q2. Computing the 2x2 covariance matrix of the sample based on sample mean
cov_matrix = np.cov(samples, rowvar=False)
print("Covariance matrix of the sample :\n", cov_matrix)
```

```python
# Q3. Finding the eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nEigenvalues of the covariance matrix :\n", eigenvalues)
print("\nEigenvectors of the covariance matrix :\n", eigenvectors)
```

```python
# Q4. Plotting Data Points with Eigen Vectors
plt.figure(figsize=(8, 8))

# Plot all 100 data points
plt.scatter(X1, X2, alpha=0.5, label='Data points')

# Plot the mean ponit
plt.scatter(*mean_sample, color='black', label='Mean')

# Plot the eigenvectors
for i in range(len(eigenvalues)):
    eig_vec = eigenvectors[:, i]
    eig_val = eigenvalues[i]
    plt.quiver(*mean_sample, eig_vec[0] * eig_val, eig_vec[1] * eig_val,
               angles='xy', scale_units='xy', scale=1, color=['r', 'b'][i], label=f'Eigenvector {i+1}')

# Set axis limits
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.axhline(0, color='grey', lw=0.75)
plt.axvline(0, color='grey', lw=0.75)
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('2D Scatter Plot of Data Points with Eigenvectors')
plt.show()
```

```python
# Q5. Rotating Data Points
# Sort the eigenvectors by eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Construct U matrix with unit eigenvectors
U = eigenvectors

# Center the sample points by subtracting the mean
centered_samples = samples - mean_sample

# Rotate each sample point by U^T
rotated_samples = centered_samples @ U.T

# Plot the rotated points
plt.figure(figsize=(8, 8))

# Plot all n=100 rotated data points
plt.scatter(rotated_samples[:, 0], rotated_samples[:, 1], alpha=0.5, label='Rotated points')

# Set axis limits
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.axhline(0, color='grey', lw=0.75)
plt.axvline(0, color='grey', lw=0.75)
plt.gca().set_aspect('equal', adjustable='box')

# Add labels and legend
plt.xlabel('X1_rotated)')
plt.ylabel('X2_rotated)')
plt.legend()
plt.title('Rotated 2D Scatter Plot of data points')
plt.show()
```
