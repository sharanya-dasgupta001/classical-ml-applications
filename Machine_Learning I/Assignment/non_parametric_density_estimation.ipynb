```python
def K(u):
  if -0.5 <= u <= 0.5 :
    return 1
  else:
    return 0
```

```python
def p(X,h):
  N = len(X)
  ls = [0]*N
  for i in range(N):
    for j in X:
      ls[i] += K((X[i]-j)/h)
    ls[i] = ls[i]/(N*h)
  return ls
```

```python
X = [-0.57,0.25,-0.08, 1.40, -1.05, -1.00, 0.37, -1.15, 0.73, 1.59]
h = 1.5
y = p(X,h)
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# Convert lists to numpy arrays
x = np.array(X)
y = np.array(y)

# define the true objective function
def objective(x, a, b, c, d, e, f):
 return (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f

popt, _ = curve_fit(objective, x, y)
# summarize the parameter values
a, b, c, d, e, f = popt

# Generate smooth curve using the fitted parameters
x_smooth = np.linspace(min(x), max(x), 100)
y_smooth = objective(x_smooth, a, b, c, d, e, f)

# Plot the data and the fitted curve
plt.scatter(x, y, label='Data')
plt.plot(x_smooth, y_smooth, 'r-', label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('p')
plt.title('Distribution for h = 1.5')

plt.grid(True)
plt.show()
```
