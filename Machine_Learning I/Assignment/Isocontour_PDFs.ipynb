```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

# Plots the isocontours of a bivariate normal distribution.
def plot_bivariate_normal(mu, sigma, title):

  # Create a grid of points
  x = np.linspace(-6, 6, 100)
  y = np.linspace(-6, 6, 100)
  X, Y = np.meshgrid(x, y)

  # Evaluate the bivariate normal distribution at each point
  pos = np.dstack((X, Y))
  Z = multivariate_normal(mu, sigma).pdf(pos)

  # Create the contour plot
  fig, ax = plt.subplots()
  CS = ax.contour(X, Y, Z)
  ax.set_title(title)
  ax.set_xlabel("X")
  ax.set_ylabel("Y")

  # Label each contour
  ax.clabel(CS, inline=1, fontsize=8)

  plt.show()
  plt.close(fig)

# Plots the isocontours of difference of two bivariate normal distributions
def plot_bivariate_normal_difference(mu1, sigma1, mu2, sigma2, title):

  # Create a grid of points
  x = np.linspace(-6, 6, 100)
  y = np.linspace(-6, 6, 100)
  X, Y = np.meshgrid(x, y)

  # Finding values of bivariate normal distribution at each point
  pos = np.dstack((X, Y))
  Z1 = multivariate_normal(mu1, sigma1).pdf(pos)
  Z2 = multivariate_normal(mu2, sigma2).pdf(pos)

  Z = Z1 - Z2

  # Create the contour plot
  fig, ax = plt.subplots()
  CS = ax.contour(X, Y, Z)
  ax.set_title(title)
  ax.set_xlabel("X")
  ax.set_ylabel("Y")
  # Label each contour
  ax.clabel(CS, inline=1, fontsize=8)

  plt.show()
  plt.close(fig)
```

```python
# Define the means and covariances for the different cases
mu1 = [1, 1]
sigma1 = [[1, 0], [0, 2]]

mu2 = [-1, 2]
sigma2 = [[2, 1], [1, 4]]

mu3_1 = [0, 2]
mu3_2 = [2, 0]
sigma3 = [[2, 1], [1, 1]]

mu4_1 = [0, 2]
mu4_2 = [2, 0]
sigma4_1 = [[2, 1], [1, 1]]
sigma4_2 = [[2, 1], [1, 4]]

mu5_1 = [1, 1]
mu5_2 = [-1, -1]
sigma5_1 = [[2, 0], [0, 1]]
sigma5_2 = [[2, 1], [1, 2]]
```

```python
# Plot the isocontours for each case
plot_bivariate_normal(mu1, sigma1, "Case 1")
plot_bivariate_normal(mu2, sigma2, "Case 2")
plot_bivariate_normal_difference(mu3_1, sigma3, mu3_2, sigma3, "Case 3")
plot_bivariate_normal_difference(mu4_1, sigma4_1, mu4_2, sigma4_2, "Case 4")
plot_bivariate_normal_difference(mu5_1, sigma5_1, mu5_2, sigma5_2, "Case 5")
```
