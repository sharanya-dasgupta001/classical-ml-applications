```python
import numpy as np

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the gradient of the cost function
def gradient(X, y, w):
    m = X.shape[0]
    s = sigmoid(np.dot(X, w))
    return np.dot(X.T, (s - y)) / m

# Define the Hessian of the cost function
def hessian(X, y, w):
    m = X.shape[0]
    s = sigmoid(np.dot(X, w))
    S = np.diag(s * (1 - s))
    return np.dot(np.dot(X.T, S), X) / m

# Perform one iteration of Newton's method
def newtons_method(X, y, w):
    grad = gradient(X, y, w)
    H = hessian(X, y, w)
    H_inv = np.linalg.inv(H)
    w_new = w - np.dot(H_inv, grad)
    return w_new
```

```python
# Define the data
X = np.array([
    [0.2, 3.1, 1.0],
    [1.0, 3.0, 1.0],
    [-0.2, 1.2, 1.0],
    [1.0, 1.1, 1.0]])

y = np.array([1, 1, 0, 0])

# Initial weight vector
w = np.array([-1, 1, 0])
```

```python
# Print initial value of s
s_initial = sigmoid(np.dot(X, w))
print(f"Initial value of s: {s_initial}")

# Perform one iteration and print the value of w
w_after_1_iter = newtons_method(X, y, w)
print(f"Value of w after 1 iteration: {w_after_1_iter}")

# Print the value of s after 1 iteration
s_after_1_iter = sigmoid(np.dot(X, w_after_1_iter))
print(f"Value of s after 1 iteration: {s_after_1_iter}")

# Perform another iteration and print the value of w
w_after_2_iters = newtons_method(X, y, w_after_1_iter)
print(f"Value of w after 2 iterations: {w_after_2_iters}")
```
